```python

# ==========================================
# BELAJAR PYTHON DASAR - INTERAKTIF
# ==========================================

print("=" * 30)
print("SELAMAT DATANG DI KELAS PYTHON")
print("=" * 30)

# --- 1. VARIABEL DAN TIPE DATA ---
# Variabel adalah tempat menyimpan data
nama = "Andi"          # Tipe data: String (Teks)
umur = 20              # Tipe data: Integer (Bilangan Bulat)
tinggi = 175.5         # Tipe data: Float (Bilangan Desimal)
is_student = True      # Tipe data: Boolean (True/False)

print("\n--- 1. Data Diri ---")
print(f"Nama: {nama}")
print(f"Umur: {umur} tahun")
print(f"Tinggi: {tinggi} cm")
print(f"Status: {'Mahasiswa' if is_student else 'Umum'}")

# --- 2. INPUT & OUTPUT (INTERAKSI USER) ---
print("\n--- 2. Kalkulator Sederhana ---")
print("Sekarang kita akan menjumlahkan dua angka.")

# Mengambil input dari user dan mengkonversi ke integer
angka1 = int(input("Masukkan angka pertama: "))
angka2 = int(input("Masukkan angka kedua: "))

# Operasi Aritmatika
hasil = angka1 + angka2
print(f"Hasil penjumlahan {angka1} + {angka2} = {hasil}")

# --- 3. KONDISI (IF / ELSE) ---
# Menggunakan logika untuk menentukan grade
print("\n--- 3. Penentuan Grade Nilai ---")
nilai = int(input("Masukkan nilai kamu (0-100): "))

if nilai >= 90:
    grade = "A"
    pesan = "Pertahankan!"
elif nilai >= 80:
    grade = "B"
    pesan = "Bagus sekali."
elif nilai >= 70:
    grade = "C"
    pesan = "Lumayan."
else:
    grade = "D"
    pesan = "Harus belajar lebih giat."

print(f"Grade kamu: {grade}")
print(f"Pesan: {pesan}")

# --- 4. PERULANGAN (LOOPS) ---
print("\n--- 4. Latihan Perulangan ---")

# Perulangan For (Mengulang jumlah tertentu)
print("Menghitung mundur dari 5:")
for i in range(5, 0, -1):
    print(i, end=" ")
print("GO!")

# Perulangan While (Mengulang selama kondisi benar)
print("\nCoba tebak angka (1-10):")
angka_rahasia = 7
tebakan = 0
percobaan = 0

while tebakan != angka_rahasia:
    tebakan = int(input("Masukkan angka: "))
    percobaan += 1
    if tebakan < angka_rahasia:
        print("Terlalu kecil!")
    elif tebakan > angka_rahasia:
        print("Terlalu besar!")

print(f"Benar! Angka其为 {angka_rahasia}. Kamu menebak dalam {percobaan} kali.")

# --- 5. LIST & DICTIONARY (STRUKTUR DATA) ---
print("\n--- 5. Struktur Data ---")

# List (Array) - Urutan indeks
buah = ["Apel", "Jeruk", "Mangga", "Pisang"]
print(f"List Buah: {buah}")
print(f"Buah kedua: {buah[1]}")

# Menambah data ke list
buah.append("Durian")
print(f"Setelah ditambahkan: {buah}")

# Dictionary (Objek Key-Value)
siswa = {
    "nama": "Budi",
    "kelas": "XII IPA 1",
    "nilai": 85
}
print(f"Data Siswa: {siswa}")
print(f"Nama Siswa: {siswa['nama']}")

# --- 6. FUNGSI ---
print("\n--- 6. Penggunaan Fungsi ---")

def salam(nama_user):
    """Fungsi untuk menampilkan salam"""
    print(f"Halo {nama_user}! Selamat belajar Python.")

def luas_persegi(sisi):
    """Fungsi untuk menghitung luas persegi"""
    return sisi * sisi

# Memanggil fungsi
salam(nama)
sisi_persegi = 4
luas = luas_persegi(sisi_persegi)
print(f"Luas persegi dengan sisi {sisi_persegi} adalah {luas}")

print("\n" + "=" * 30)
print("\n" + " " * 3, "SEMOGA BERMANFAAT!")
print("=" * 30)
```

```python
import random

word = ["Python", "Kucing", "Komputer", "Logika", "Algoritma"]

def scramble(word):
    letters = list(word)
    random.shuffle(letters)

    # TODO: gabungkan kembali huruf menjadi string
    scrambled = "".join(letters)

    return scrambled


secret_word = random.choice(word)
scrambled_word = scramble(secret_word)

print("Tebak kata ini:")
print(scrambled_word)

attempts = 3

while attempts > 0:
    guess = input("Tebakanmu: ")

    # TODO: cek apakah tebakan benar
    if guess == secret_word:
        print("Benar!")
        break
    else:
        attempts -= 1
        print("Salah. Sisa percobaan:", attempts)


# TODO: jika percobaan habis, beri tahu jawabannya
else: print("percobaan habis jawabannya =", secret_word)
```
